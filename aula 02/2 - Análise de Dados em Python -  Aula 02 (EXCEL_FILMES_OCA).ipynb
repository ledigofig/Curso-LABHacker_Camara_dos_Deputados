```python
#Tirando o fim com (skipfooter)
df = pd.read_excel('src/2120.xlsx', skiprows=2, skipfooter=16)

#Exibindo os primeiros dados
df.head()
```

```python
df.tail()
```

```python
# Veriicar todos os tipos das colunas 
df.dtypes
# Não reconheceu o 'Data de Lançamento' como data
```

```python
# Convertendo o campo data de lançamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# indentificando o problema da coluna data de lançamento
df['Data de lançamento'].value_counts()
# Problema 
# Relançamento           175
```

```python
#vendo quantidade de linhas e colunas
df.shape
```

```python
#Criando dataframe novo
df_novo = df[df['Data de lançamento' ] != 'Relançamento'].copy()

df_novo.head()
```

```python
df_novo.dtypes
```

```python
# Convertendo o campo data de lançamento em data
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
#Quas os anos tiveram mais filmes lançados?

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
#Mostrando em grafico oos dados de lançamento por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar(figsize=(16,8))
```

```python
# Qual o filme teve maior bilheteria 
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()].
```

```python
# Vendo top 10 de bilheteria de filmes 
df_novo[['Ano de exibição','Título da obra','Renda (R$) no ano de exibição']].nlargest(10, 'Renda (R$) no ano de exibição')
```

```python
# Vendo top 10 de bilheteria de filmes 
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
# QUANTOS FILMES BRASILEIROS E ESTRANGEIROS
df_novo['Nacionalidade da obra'].value_counts()
```
