A fonte de dados Ã© :http://educacao.dadosabertosbr.com/

```python
#importando as bibliotecas 
import pandas as pd
import requests

#Url da api
url_api = 'http://educacao.dadosabertosbr.com/api/escolas/buscaavancada?situacaoFuncionamento=1&energiaInexistente=on&aguaInexistente=on&esgotoInexistente=on&cozinha=on'

#RequisitANDO api COM REQUESTS
resposta =requests.get(url_api)


#criando dataframe
df = pd.DataFrame(resposta.json()[1])
df.head()
```
