```python
#importando
import pandas as pd
%matplotlib inline 
```

```python
#carrengando dados
url_curso_prouni = 'https://raw.githubusercontent.com/rodolfoghi/curso-labhacker-analise-dados-com-python/master/cursos-prouni.csv'

df = pd.read_csv(url_curso_prouni)
df.head()
```

```python
#verificando tamanho da dataframe
# Formato do shape(linha , Colunas)
df.shape
```

```python
#Tipos de dados das colunas
df.dtypes
```

```python
# Analise matematica das colunas numericas
df.describe()
```

```python
# Decreve dados
df.info()
```

```python
# Quantidade de curso por UF
df['uf_busca'].value_counts()
```

```python
df['uf_busca'].value_counts(ascending = True ).plot.barh(
    title='Quantidade de cursos por UF', figsize = (20, 12) )
```

```python
#qual o curso com mensalidade mais alta?
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
#qual o curso com mensalidade mais barato?
df[df['mensalidade'] == df['mensalidade'].min()]
```

```python
# Qual a mensalidade mais cara por UF
df.groupby('uf_busca')['mensalidade'].max()
```

```python
df.groupby('uf_busca')['mensalidade'].max().plot.barh(
    title='Quantidade de cursos por UF', figsize = (20, 12) )
```

```python
# Qual a mensalidade mais barata por UF
df.groupby('uf_busca')['mensalidade'].min()
```

```python
# mostrando todos os dados
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```python
# Qual a mensalidade 500 e 1000
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
#quais as 5 maiores mensalidades
df.nlargest(5, 'mensalidade')
```

```python
#quais as 5 menores mensalidades
df.nsmallest(6, 'mensalidade')
```
