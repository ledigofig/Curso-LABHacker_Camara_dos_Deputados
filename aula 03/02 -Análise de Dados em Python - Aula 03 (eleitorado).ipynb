```python
#importando as bibliotecas 
import pandas as pd
import seaborn
```

```python
df = pd.read_csv('src/eleitorado_municipio_2018.csv', encoding='latin1', sep=';')

df.head().T
```

```python
df.info()
```

```python
# As 5 cidades com maior quantidade de eleitores  deficiente
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('eleitoras :', df['QTD_ELEITORES_FEMININO'].sum())
print('eleitores :', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não informados :', df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
#criação de variaveis
total_geral = df['QTD_ELEITORES'].sum()
total_eleitoras = df['QTD_ELEITORES_FEMININO'].sum()
total_eleitores = df['QTD_ELEITORES_MASCULINO'].sum()
total_nao_informados = df['QTD_ELEITORES_NAOINFORMADO'].sum()

```

```python
print('eleitoras :', ((total_eleitoras / total_geral)*100).round(2))
print('eleitores :',  ((total_eleitores / total_geral)*100).round(2))
print('Não informados :',  ((total_nao_informados / total_geral)*100).round(2))
```

```python
#quantos municipios com mais homens que mulheres
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# Criando uma coluna nova para indicar a relação fem/masc
df['relacao_fm'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
# Quais municipios com maior relação f/m
df.nlargest(10,'relacao_fm').T
```

```python
# Criando um dataframe só com dados do Brasil

df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head().T
```

```python
#municipios com maior relação 
df_brasil.nlargest(10,'relacao_fm').T
```

```python
#municipios com maior relação 
df_brasil.nsmallest(10,'relacao_fm').T
```

```python
# importando seaborn 
#essa biblioteca para desenhar grafico
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Embelezando o label abaixo do texto

sns.distplot(df_brasil['relacao_fm'], bins=100, color='red', kde=False)

plt.title('Relação Eleitores/eleitora', fontsize =18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
```

```python
# Embelezando o grafico colocando lebal y com nome frequencia 

sns.distplot(df_brasil['relacao_fm'], bins=100, color='red', kde=False)

plt.title('Relação Eleitores/eleitora', fontsize =18)
plt.xlabel('Eleitoras/Eleitores', fontsize=14)
plt.ylabel('Frequencia', fontsize=14)
```

```python
# Embelezando o grafico colocando a divisão no meio com pontilhado

sns.distplot(df_brasil['relacao_fm'], bins=100, color='red', kde=False)

plt.title('Relação Eleitores/eleitora', fontsize =18)
plt.xlabel('Eleitores/Eleitoras', fontsize=14)
plt.ylabel('Frequencia', fontsize=14)
plt.axvline(1.0, color='black',linestyle='--')
```
