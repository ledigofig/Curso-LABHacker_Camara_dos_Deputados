```python
#voltando a carga para separador de decimal
df = pd.read_csv('src/2018.csv',sep=';', encoding='latin1', skiprows=1, decimal=',')

df.head()
```

```python
#verificando tamnho e tipo dos dados
df.info()
```

```python
df.describe()
```

**Qual o total de reembolso?**

```python
df['VALOR_REEMBOLSADO'].sum()
```

```python
# Quantos reembolsos foram solicitados por cada senador
df['SENADOR'].value_counts()
```

```python
# Vendo o total que cada senador solicitou de reembolso
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
# Quais são os 5 maiores valore de rembolso?
# .T TRASCREVE A FORMA DE VISUALIZAR OS DADOS
df.nlargest(5, 'VALOR_REEMBOLSADO').T
```

```python
#sabendo os 5 menores rembolsos
df.nsmallest(5, 'VALOR_REEMBOLSADO').T
```

```python
# Quantidade de rembolso por tipo de despesa
df['TIPO_DESPESA'].value_counts()
```

```python
#Valores totais por tipo de despesa
df.groupby('TIPO_DESPESA')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```
